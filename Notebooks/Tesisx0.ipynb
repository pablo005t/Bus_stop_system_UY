```python
import geopandas as gpd


from google.colab import drive
drive.mount('/content/drive')

# Assuming the file is a GeoJSON in your Google Drive
# You might need to adjust the path based on where your file is located
paradas_omnibus_path = '/content/drive/My Drive/Tesis/Datos/Paradas_omnibus_UY.geojson'

paradas_omnibus = gpd.read_file(paradas_omnibus_path)
```

```python
import requests
import time
from urllib.parse import urlencode
import os
from google.colab import userdata


API_KEY = userdata.get('GOOGLE_MAPS_API_KEY')

if not API_KEY:
    raise RuntimeError("No se encontró la variable de entorno GOOGLE_API_KEY en Colab secrets.")


def pedir_imagen(lat, lon, zoom=20, size=(640, 640), scale=1, maptype="satellite", retries=3, sleep_base=1):
    """
    Descarga una imagen estática de Google Maps centrada en lat, lon.
    Devuelve el contenido en bytes.
    """
    params = {
        "center": f"{lat},{lon}",
        "zoom": zoom,
        "size": f"{size[0]}x{size[1]}",
        "scale": scale,
        "maptype": maptype,
        "key": API_KEY
    }
    url = "https://maps.googleapis.com/maps/api/staticmap?" + urlencode(params)

    attempt = 0
    while attempt < retries:
        resp = requests.get(url, timeout=10)
        if resp.status_code == 200 and resp.headers.get("Content-Type", "").startswith("image"):
            return resp.content
        else:
            attempt += 1
            wait = sleep_base * (2 ** (attempt - 1))
            print(f"Intento {attempt} fallido (status {resp.status_code}), reintentando en {wait}s...")
            time.sleep(wait)
    raise RuntimeError(f"No se pudo descargar imagen después de {retries} intentos. Último status: {resp.status_code}")
```

```python



output_dir = '/content/drive/My Drive/Tesis/Datos/Imagenes_Satelitales'
os.makedirs(output_dir, exist_ok=True)

# Solo zoom 20
zoom = 20

for idx, row in paradas_omnibus.iterrows():
    lat = row.geometry.y
    lon = row.geometry.x
    try:
        print(f"Descargando parada {idx}, zoom {zoom}...")
        img_bytes = pedir_imagen(lat, lon, zoom=zoom)
        filename = f"parada_{idx}_zoom_{zoom}.png"
        filepath = os.path.join(output_dir, filename)
        with open(filepath, "wb") as f:
            f.write(img_bytes)
        print(f"  Guardado en: {filepath}")
    except Exception as e:
        print(f"  Error parada {idx} zoom {zoom}: {e}")
```

```python
# Instala el SDK (descomenta si aún no lo tienes)
# !pip install inference_sdk pillow

from inference_sdk import InferenceHTTPClient
from google.colab import drive, userdata
from pathlib import Path
from PIL import Image
import json

# 1) Montar Google Drive
drive.mount('/content/drive')

# 2) Cargar tu Roboflow Private API Key
API_KEY = userdata.get('Roboflow_API_Key')
if not API_KEY:
    API_KEY = input("🔑 Introduce tu Roboflow Private API Key: ").strip()
if not API_KEY:
    raise RuntimeError("Se necesita la Roboflow API Key para continuar.")

# 3) Inicializar el cliente
client = InferenceHTTPClient(
    api_url="https://serverless.roboflow.com",
    api_key=API_KEY
)


# 4) Configuración de rutas y pipeline
WORKSPACE    = "tesis-um-n0pxr"           # tu workspace
WORKFLOW_ID  = "detect-and-visualize-2"   # el ID de tu pipeline desplegado
UNLABELED    = Path('/content/drive/My Drive/Tesis/Datos/Imagenes_Satelitales')
OUTPUT_DIR   = Path('/content/drive/My Drive/Tesis/Datos/predicciones_sugeridas')
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# 5) Bucle de inferencia y guardado
for img_path in UNLABELED.glob("*.png"):
    try:
        # Cargar imagen en PIL y convertir a RGB para JPEG interno
        img = Image.open(img_path).convert("RGB")

        # Ejecutar workflow
        result = client.run_workflow(
            workspace_name='TesisUM',
            workflow_id='pablo-vitku/detect-and-classify-4',
            images={"image": img},    # key “image” según tu Inputs
            use_cache=True
        )

        # Guardar solo el array de predictions en JSON
        out_file = OUTPUT_DIR / f"{img_path.stem}.json"
        with open(out_file, "w") as f:
            json.dump(result.predictions, f, indent=2)

        print(f"✅ Predicciones guardadas para {img_path.name}")

    except Exception as e:
        print(f"❌ Error procesando {img_path.name}: {e}")
```
